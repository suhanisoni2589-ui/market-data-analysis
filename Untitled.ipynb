# Data Collection Function

This function downloads OHLCV data and performs:

1. Multi-index flattening
2. Missing value handling
3. Date alignment preparation

```python
def download_stock_data(tickers, start_date, end_date):

    stock_data = {}

    for label, ticker in tickers.items():

        df = yf.download(
            ticker,
            start=start_date,
            end=end_date,
            auto_adjust=False
        )

        if isinstance(df.columns, pd.MultiIndex):
            df.columns = df.columns.get_level_values(0)

        df = df.ffill()

        stock_data[label] = df

        print(
            f"{label}: shape={df.shape}, "
            f"{df.index.min().date()} to {df.index.max().date()}"
        )

    return stock_data
```

```python
stock_data = download_stock_data(
    tickers,
    start_date,
    end_date
)
```

# Part A — Systematic Risk: Beta Estimation

Beta measures the sensitivity of a stock's return relative to market movements.

Interpretation:

- Beta > 1 → Aggressive
- Beta = 1 → Market-like
- Beta < 1 → Defensive

Formula:

β = Cov(Ri,Rm) / Var(Rm)

## Calculating Daily Returns

Daily percentage returns are calculated using:

rt = (Pt − Pt−1)/Pt−1

The pct_change() method is used for consistency.

```python
returns_df = pd.DataFrame()

for label, df in stock_data.items():
    returns_df[label] = df["Adj Close"].pct_change()

market = yf.download(
    market_ticker,
    start=start_date,
    end=end_date
)

returns_df["NIFTY50"] = market["Adj Close"].pct_change()

returns_df = returns_df.dropna()

returns_df.head()
```
